## PCA, KMeans Clustering, and K Nearest Neighbors

## Scale your data.

```python
# Your code here:
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
dataset_scaled = scaler.fit_transform(dataset)
```

## Use the first two `PCA` components to visualize your data.

* Use your **target** as your color parameter.

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=270)
dataset_scaled_pca = pca.fit_transform(dataset_scaled)
plt.scatter(dataset_scaled_pca[:,0],dataset_scaled_pca[:,1], c=target,cmap='magma')
plt.colorbar()
```

## Create a scree plot by plotting the explained variance and cumulative explained variance against component number.

```python
explained_var_ratio_cum = np.cumsum(pca.explained_variance_ratio_)
index=len(dataset.columns)
```

```python
# Your code here:
plt.plot(explained_var_ratio_cum)
plt.bar(range(0,270),pca.explained_variance_ratio_)
plt.title('cumulative explained variance')
plt.xlabel('component')


plt.ylabel('explained variance',color='b')

```

## How many components do you need to explain 90% of your variance?

```python
np.where(np.cumsum(pca.explained_variance_ratio_) > .9)[0][0]
```

## Use KMeans to cluster your data. Evaluate different values of K visually and with the silhouette score. Justify your choice.

```python
ks = []
sils = []
plt.figure(figsize=(16,20))
for i,k  in enumerate(range(2,8)):
    km=KMeans(n_clusters=k, random_state=42)
    km.fit(dataset_scaled)
    plt.subplot(321+i)
    plt.scatter(dataset_scaled_pca[:,0], dataset_scaled_pca[:,1], c=km.labels_, cmap='magma')
    plt.xlabel('PC1')
    plt.ylabel('PC2') 
    plt.title('K:{} \nsil:{}'.\
              format(k, round(silhouette_score(dataset_scaled_pca, km.labels_),4)))
    
    plt.colorbar()
    ks.append(k)
    sils.append(silhouette_score(dataset_scaled_pca, km.labels_))
```

## Repeat the step above, excpt use `silhouette_samples` to make the color of each point its silhouette score. Which clusters are the best defined? Are there any that are more poorly defined?

#### what is the strong cluster mean?

```python
plt.figure(figsize=(16,20))
for i,k  in enumerate(range(2,8)):
    km=KMeans(n_clusters=k, random_state=42)
    km.fit(dataset_scaled)
    sil = silhouette_score(dataset_scaled_pca,km.labels_)
    plt.subplot(321+i)
    plt.scatter(dataset_scaled_pca[:,0], dataset_scaled_pca[:,1], \
                c=silhouette_samples(dataset_scaled_pca, km.labels_), cmap='magma')
    plt.xlabel('PC1')
    plt.ylabel('PC2') 
    plt.title('K:{}\nsil:{}'.
              format(k,round(sil,4)))
    
    plt.colorbar()
```

#### ANSWER: 
            ** Although the silhouette score is higher on for K = 2 **
            ** I think clusters are best defined with KNN n_neighbors = 3 on the PC1/ PC2 plots by visualizatio. **
                  

```python
plt.plot(ks,sils)
plt.ylim(-.1,.1)
plt.xlim(1,8)
plt.xlabel('K')
plt.ylabel('Silhouette Score')
```

## Create a temporary target, `km_target`, from the `KMeans.labels_` from your KMeans clustering.

```python
km_target=km.labels_
km_target
```

## Use `KNeighborsClassifier` to build a model predicting your `km_target`. 

* Train test split your data
* Use GridSearchCV to cross-validate and find a good `n_neighbors` paramater.
* Evaluate your model performance using the `.score()` method, which returns accuracy.

```python
# Your code here:
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split, GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(dataset_scaled,
                                                    km_target,
                                                    test_size = .3,
                                                    random_state = 42)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
params={
    'n_neighbors':list(range(1,31,2))
}
gs_knc = GridSearchCV(KNeighborsClassifier(), param_grid=params, cv=5, n_jobs=-1)
gs_knc.fit(X_train_scaled, y_train)
```

```python
gs_knc.best_estimator_
```

```python
print('Train score: {}'.format(gs_knc.best_estimator_.score(X_train_scaled, y_train)))
print('Test score: {}'.format(gs_knc.best_estimator_.score(X_test_scaled, y_test)))
```

```python
print('Train score: {}'.format(gs_knc.score(X_train_scaled, y_train)))
print('Test score: {}'.format(gs_knc.score(X_test_scaled, y_test)))
```

## Plot your train and test scores from your `GridSearchCV.cv_results_` object over different values of k. Did this method yield a good value for k? 

#### high bias --> low train score (model unable to learn from data)
#### high variance ---> low test score (model unable to generalize)
#### Cross-validation help us identify bias and variance by siulating multiple train-test splits and evaluating model on those splits
#### Grid-search helps us reduce the bias and variance of our model by tuning hyperparamters without test data

```python
plt.plot(params['n_neighbors'],gs_knc.cv_results_['mean_train_score'],label='train')
plt.plot(params['n_neighbors'],gs_knc.cv_results_['mean_test_score'],label='test')
plt.axvline(gs_knc.best_estimator_.n_neighbors, c='g', label='best k')
plt.legend()
```

## Evaluate your model. Does it have high bias? High variance? How does GridSearchCV help reduce bias and variance?

```python
## this model does not have high bias nor high variance
## GridSearchCV helped reduce biace & variance

print(gs_knc.best_estimator_)
print('Train score: {}'.format(gs_knc.score(X_train_scaled, y_train)))
print('Test score: {}'.format(gs_knc.score(X_test_scaled, y_test)))
```
